```python
import sys
import os

PROJECT_ROOT = os.path.abspath('..') 

if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

from src.data_features.data_loader import load_data, describe_data

data_path = os.path.join(PROJECT_ROOT, 'data', 'creditcard.csv')
print(f"Looking for data at: {data_path}") 

df = load_data(data_path)
describe_data(df)
```

```python
df[df["Class"] == 1].iloc[0]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

class_counts = df['Class'].value_counts()
class_counts
```

```python
plt.figure(figsize=(12,12))
sns.countplot(data=df,x='Class')
plt.title('Class Distribution')
plt.xlabel('Class (0 = Legitimate, 1 = Fraud)')
plt.ylabel('Count')

plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.histplot(df['Class'],stat='percent',discrete=True)
plt.title('Class Distribution (%)')
plt.xlabel('Class (0 = Legit, 1 = Fraud)')
plt.ylabel('Percentage')
plt.show()
```

Amount distribution for valid transactions

```python
legit = df[df['Class']==0]
sns.histplot(legit['Amount'],bins=100,log_scale=True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df,x='Amount',hue='Class',bins=100,log_scale=True)
plt.title('Amount distribution , log scale')
```

```python
sns.boxplot(data=df,x='Class',y='Amount')
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(data=df, x='Time', hue='Class',
             bins=100, kde=True,
             palette=['#0D7377', '#D4820A'],
             alpha=0.5)

plt.title('Transaction Timing Distribution')
plt.show()
```
